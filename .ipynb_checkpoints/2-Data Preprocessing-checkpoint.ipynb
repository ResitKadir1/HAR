```python
# get the features from the file features.txt

with open('UCI_HAR_Dataset/features.txt') as f:
    features =[r.strip().split()[-1] for r in f]
with open('UCI_HAR_dataset/activity_labels.txt') as f:
    labels = [r.strip().split()[-1] for r in f]

print('No of Features: {}'.format(len(features)))
print('No of labels: {}'.format(len(labels)))
```

```python
labels
```

```python
features
```

## Train Dataset

```python
X_train = pd.read_csv('UCI_HAR_dataset/train/X_train.txt',
                      delim_whitespace=True, header=None)
X_train.columns = features
features
```

```python
# add subject column to the dataframe
X_train['subject'] = pd.read_csv('UCI_HAR_dataset/train/subject_train.txt',
                                 header=None, squeeze=True)
```

```python
X_train.sample(10)
```

```python
y_train = pd.read_csv('UCI_HAR_dataset/train/y_train.txt', names=['Activity'], squeeze=True)
y_train_labels = y_train.map({1: 'WALKING', 2:'WALKING_UPSTAIRS',3:'WALKING_DOWNSTAIRS',\
                       4:'SITTING', 5:'STANDING',6:'LAYING'})
print(y_train.sample(4))
print(y_train_labels.sample(4))
```

```python
# put all columns in a single dataframe
train = X_train
train['Activity'] = y_train
train['ActivityName'] = y_train_labels
train.sample(5)
```

```python
train.shape
```

## Test Dataset

```python
X_test = pd.read_csv('UCI_HAR_dataset/test/X_test.txt', delim_whitespace=True, header=None)
X_test.columns = features
features
X_test.sample(10)
```

```python
# add subject column to the dataframe
X_test['subject'] = pd.read_csv('UCI_HAR_dataset/test/subject_test.txt',
                                header=None, squeeze=True)

X_test.sample(10)
```

```python
# get y labels from the txt file
y_test = pd.read_csv('UCI_HAR_dataset/test/y_test.txt', names=['Activity'], squeeze=True)
y_test_labels = y_test.map({1: 'WALKING', 2:'WALKING_UPSTAIRS',3:'WALKING_DOWNSTAIRS',\
                       4:'SITTING', 5:'STANDING',6:'LAYING'})
print(y_test.head(2))
print(y_test_labels.head(2))
```

```python
# put all columns in a single dataframe
test = X_test
test['Activity'] = y_test
test['ActivityName'] = y_test_labels
test.sample(4)
```

```python
test.shape
```

# Data Cleaning

```python
## 1. Check for Duplicates
print('No of duplicates in train: {}'.format(sum(train.duplicated())))
print('No of duplicates in test : {}'.format(sum(test.duplicated())))
## 2. Checking for NaN/null values
print('We have {} NaN/Null values in train'.format(train.isnull().values.sum()))
print('We have {} NaN/Null values in test'.format(test.isnull().values.sum()))
```

### 1.  Check for data imbalance

```python
sns.set_style('whitegrid')
plt.rcParams['font.family'] = 'Dejavu Sans'
```

```python

if not os.path.exists('fig'):
    os.makedirs('fig')
fig = "fig/"
```

```python
plt.figure(figsize=(20,10))
plt.title('Data provided by each user', fontsize=30)
sns.countplot(x='subject',hue='ActivityName', data = train)
plt.savefig(fig+'Data provided by each user.png', bbox_inches='tight')
plt.show()
```

__We have got almost same number of reading from all the subjects__

```python
plt.figure(figsize=(20,10))
plt.title('No of Datapoints per Activity', fontsize=15)

sns.countplot(train.ActivityName)
plt.xticks(rotation=45)
plt.savefig(fig+'No of Datapoints per Activity.png', bbox_inches='tight')
plt.show()
```

### Observation
> Our data is well balanced (almost)

### 2. Rename feature names 

```python
train.columns
```

```python
def clean_col(col):
    # Removing '()' from column names
    col = col.str.replace('[()]','')
    col = col.str.replace('[-]', '')
    col = col.str.replace('[,]','')
    return col
```

```python
train.columns = clean_col(train.columns)
test.columns  = clean_col(test.columns)
```
